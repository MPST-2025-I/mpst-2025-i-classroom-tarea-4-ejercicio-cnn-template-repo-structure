Este paper propone un nuevo modelo de arquitectura para la segmentacion semantica en entornos no estructurados. El modelo propuesto es una mejora del modelo UNet, que es un modelo de segmentacion semantica muy popular que incluye convoluciones y UpConvolutions que hacen lo contrario de las convoluciones, es decir, que aumentan el tamaño de la imagen.

Lo que hace UNET es que tiene una arquitectura en forma de U, que tiene una parte de convoluciones y otra de UpConvolutions. La parte de convoluciones es la que va a reducir el tamaño de la imagen, y la parte de UpConvolutions es la que va a aumentar el tamaño de la imagen. Con las excepciones que algunas imagenes saltan hasta la expansion sin pasar por la convolucion(Skip connections). Sin embargo, el modelo propuesto por este paper, propone una mejora de este modelo, que es el Eff-UNet.


 MBConv escala uniformemente la profundidad, el ancho y la resolución de la red para mejorar el rendimiento con un conjunto fijo de factores de escala
   - Ancho: Neuronas en cada capa
   - Profundidad: Número de capas
   - Resolución: Tamaño de la imagen
   - Compuesto: Todos juntos
